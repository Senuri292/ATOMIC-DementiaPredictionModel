```python
# function to find outliers
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data < (q1 - 1.5 * IQR)) | (data > (q3 + 1.5 * IQR)))]

  return outliers
```

```python
# Finding outliers for anything
outliers = find_outliers_IQR(data['anything'])

print("number of outliers: "+ str(len(outliers)))

outliers
```

```python
# Fuzzy matching for typos
from rapidfuzz import process, fuzz
choices = data['city'].unique()
def fix_typo(x):
    best_match, score, _ = process.extractOne(x, choices, scorer=fuzz.token_sort_ratio)
    return best_match if score > 80 else x

data['city'] = data['city'].apply(fix_typo)
```
