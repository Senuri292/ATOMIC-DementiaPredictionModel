```python
# function to find outliers
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data < (q1 - 1.5 * IQR)) | (data > (q3 + 1.5 * IQR)))]

  return outliers
```

```python
# Finding outliers for anything
outliers = find_outliers_IQR(data['NACCDAYS'])

print("number of outliers: "+ str(len(outliers)))

outliers
```

```python
outlier_rows = data.index[data["NACCDAYS"] > 5000].tolist()
print(len(outlier_rows))
print(outlier_rows)
```

Feature Selection Part 2

```python
# remove columns that has almost the same value for all rows(low variance)
# remove if variance is less than 0.01.

from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0.01)
data_reduced = selector.fit_transform(data)

# Keep only the columns that were not removed
selected_columns = data.columns[selector.get_support()]

data = pd.DataFrame(data_reduced, columns=selected_columns)
```

```python
y = data['DEMENTED']  
X = data.drop(columns=['DEMENTED'])

from sklearn.feature_selection import mutual_info_classif

mi = mutual_info_classif(X, y)
mi_series = pd.Series(mi, index=X.columns)
selected_one= mi_series.sort_values(ascending=False).head(50)
print(selected_one)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(data, data['DEMENTED'])

importances = pd.Series(model.feature_importances_, index=data.columns)
selected_two = importances.sort_values(ascending=False).head(50)
print(selected_two)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.01)
lasso.fit(data, data['DEMENTED'])

selected_three = data.columns[lasso.coef_ != 0]
print(selected_three)
```

```python
final_features = set(selected_one.index) | set(selected_two.index) | set(selected_three)
final_features_list = sorted(list(final_features))
print(final_features_list)
print(len(final_features_list))
```

```python
final_data = data[final_features_list]
print(f"Original shape: {data.shape}")
print(f"Selected shape: {final_data.shape}")
print(f"Features kept: {len(final_features_list)}")
```

```python
# detect categorical columns
categorical_cols = [col for col in final_data.columns if final_data[col].nunique() <= 20 and "MO" not in col.upper() and "DAY" not in col.upper() and "DY" not in col.upper()]

# replace 9 and 99 only in categorical columns
final_data[categorical_cols] = final_data[categorical_cols].replace([9, 99, -4], np.nan)

#fill NaN with mode
final_data[categorical_cols] = final_data[categorical_cols].apply(lambda col: col.fillna(col.mode()[0]))

for col in categorical_cols:
    final_data[col] = final_data[col].astype(int)

encoder = OrdinalEncoder()
final_data[categorical_cols] = encoder.fit_transform(final_data[categorical_cols]) + 1
```
