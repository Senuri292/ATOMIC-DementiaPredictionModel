```python
# function to find outliers
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data < (q1 - 1.5 * IQR)) | (data > (q3 + 1.5 * IQR)))]

  return outliers
```

```python
# Finding outliers for anything
outliers = find_outliers_IQR(data['NACCDAYS'])

print("number of outliers: "+ str(len(outliers)))

outliers
```

```python
outlier_rows = data.index[data["NACCDAYS"] > 5000].tolist()
print(len(outlier_rows))
print(outlier_rows)
```

Standardising

```python
data.head(5)
```

```python
# replacing any1 and any2 string values with numerical values while making everything lowercase so no case sensitivity 
data['PACKET'] = data['PACKET'].map({'I': 1, 'F': 2, 'T': 3, 'IT': 4})
```

```python
# detect categorical columns
categorical_cols = [col for col in data.columns if data[col].nunique() <= 20 and "MO" not in col.upper() and "DAY" not in col.upper() and "DY" not in col.upper()]

# replace 9 and 99 only in categorical columns
data[categorical_cols] = data[categorical_cols].replace([9, 99, -4], np.nan)

#fill NaN with mode
data[categorical_cols] = data[categorical_cols].apply(lambda col: col.fillna(col.mode()[0]))

for col in categorical_cols:
    data[col] = data[col].astype(int)

encoder = OrdinalEncoder()
data[categorical_cols] = encoder.fit_transform(data[categorical_cols]) + 1
```
